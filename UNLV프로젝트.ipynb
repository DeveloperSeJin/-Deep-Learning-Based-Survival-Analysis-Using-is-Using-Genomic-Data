```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

data1 = pd.read_csv('./drive/MyDrive/UNLV_project/data_RNA_Seq_v2_mRNA_median_all_sample_Zscores.txt', sep='\t')
data2 = pd.read_csv('./drive/MyDrive/UNLV_project/data_bcr_clinical_data_patient.txt', sep='\t')
```

```python
data1 = data1.set_index('Hugo_Symbol', drop = True)
```

```python
data1 = data1.T
data1.rename(index = lambda x : str(x).replace('-01', ''), inplace = True)
data1 = data1.iloc[1:]
data1
```

```python
data2 = data2.iloc[4:]
data2 = data2[['Patient Identifier', 'Overall Survival Status', 'Overall Survival (Months)']]
data2 = data2.set_index('Patient Identifier', drop = True)
data2
```

```python
data = data2.merge(data1, left_on = 'Patient Identifier', right_index = True)
```

```python
data.iloc[0:10]
```
